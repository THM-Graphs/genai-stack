```python
from neo4j import GraphDatabase
from dotenv import load_dotenv
import os

load_dotenv()
uri = os.getenv("NEO4J_URI")
uri = "bolt://localhost:7687"
user = os.getenv("NEO4J_USERNAME")
passwd = os.getenv("NEO4J_PASSWORD")

with GraphDatabase.driver(uri, auth=(user, passwd)) as driver:

    # use a separate label for nodes carrying embeddings
    driver.execute_query("match (t:Text{type:'abstract'}) set t:Abstract")

    results, summary, keys = driver.execute_query("""
match (t:Text{type:'abstract'})
return t.guid as guid, t.text as text
""")
    
    for record in results:
        print(record['guid'], record['text'])

```
